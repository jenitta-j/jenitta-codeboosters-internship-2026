```
import pandas as pd
```

```
df = pd.read_csv('student_performance.csv')
```

```
df.head()
```

```
df.tail()
```

```
print("shape of dataset:", df.shape)
print("Number of students:", df.shape[0])
print("Number of columns:", df.shape[1])
print("Column in my dataset:")
print(df.columns.tolist())
```

```
print("Data types :")
print(df.dtypes)
```

```
print("Missing values in each column:")
print(df.isnull().sum())

print("\nTotal Missing values:")
print(df.isnull().sum().sum())
```

```
print("student names:")
print(df['name'])
```

```
print("Department in the dataset:")
print(df['department'].unique())
```

```
print("How many students per Department: ")
print(df['department'].value_counts())
```

```
cs_students = df[df['department'] == 'Computer Science']

print(f"Number of Computer Science students: {len(cs_students)}")

print("\nDetails of Computer Science students:")

print(cs_students[['name', 'math_score', 'programming_score']])
```

```
df['total_score']=df['math_score']+df['science_score']+df['programming_score']+df['english_score']

top_student=df.loc[df['total_score'].idxmax()]

print("Top student : ")
print(f"Name : {top_student['name']}")
print(f"Department : {top_student['department']}")
print(f"Total Score : {top_student['total_score']}")
print(f"Attendance : {top_student['attendance_percentage']}")
```

```
df_sorted=df.sort_values(by='total_score',ascending=False)
print("top 10 student:")
print(df_sorted[['name','department','total_score','attendance_percentage']].head(10))
```
